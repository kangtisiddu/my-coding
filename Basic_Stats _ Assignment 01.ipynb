```python
df = pd.read_csv(r"C:\Users\sidduling\Downloads\Sales_Dataset_Basic Stats 1.csv")
df
```

```python
numerical_cols = df.select_dtypes(include=[np.number])
```

```python
desc_stats = numerical_cols.describe().T[['mean','50%','std']]
desc_stats.rename(columns={'50%':'median'},inplace=True)
```

```python
desc_stats['mode']=numerical_cols.mode().iloc[0]
print(desc_stats)
```

```python
#histograms
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for col in numerical_cols.columns:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col],kde=True)
    plt.title(f'Histogram of {col}')
    plt.show()
```

```python
##Boxplots
for col in numerical_cols.columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
categorical_cols = df.select_dtypes(include='object')
```

```python
for col in categorical_cols.columns:
    plt.figure(figsize=(6,4))
    
```

```python
df[col].value_counts().plot(kind='bar')
plt.title(f'Bar Chart of {col}')
plt.ylabel("Frequency")
plt.show()
```

```python
standarized_df = (numerical_cols-numerical_cols.mean())/numerical_cols.std()
```

```python
print("Original mean:\n", numerical_cols.mean())
print("Standardized mean:\n", standarized_df.mean())
print("Original std:\n", numerical_cols.std())
print("Standardized std:\n", standarized_df.std())
```
