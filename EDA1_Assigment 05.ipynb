```python
print("\nMissing Values:\n", df.isnull().sum())
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
df = df.dropna()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.title("Boxplot to Detect Outliers")
plt.show()
```

```python
summary = df.describe().T
```

```python
summary['IQR'] = summary['75%'] - summary['25%']
```

```python
summary[['mean', '50%', 'std', 'IQR']]
```

```python
print("Skewness:\n", df.skew())
```

```python
print("\nKurtosis:\n", df.kurtosis())
```

```python
df.hist(figsize=(15, 10), bins=30)
plt.suptitle("Histograms of Features")
plt.tight_layout()
plt.show()
```

```python
for col in df.columns:
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[['LB', 'AC', 'FM', 'UC', 'DL']])
plt.suptitle("Pairplot (Partial Features)", y=1.02)
plt.show()
```

```python
for col in ['ASTV', 'MSTV', 'ALTV', 'MLTV']:
    sns.violinplot(x=df[col])
    plt.title(f"Violin Plot of {col}")
    plt.show()
```

```python
sns.scatterplot(x='FM', y='AC', data=df)
plt.title("Fetal Movement vs Accelerations")
plt.show()
```

```python
sns.scatterplot(x='UC', y='DL', data=df)
plt.title("UC vs DL")
plt.show()
```
