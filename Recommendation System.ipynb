```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
tfidf
```

```python
tfidf_matrix = tfidf.fit_transform(df['genre'])
```

```python
tfidf_matrix
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Normalize ratings
scaler = MinMaxScaler()
```

```python
scaler
```

```python
rating_scaled = scaler.fit_transform(df[['rating']])
```

```python
rating_scaled
```

```python
import scipy.sparse as sp
```

```python
combined_features = sp.hstack([tfidf_matrix, rating_scaled])
```

```python
combined_features
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim = cosine_similarity(combined_features, combined_features)
```

```python
cosine_sim
```

```python
# Function to recommend anime
def recommend_anime(title, top_n=5):
    # Find anime index
    idx = df[df['name'] == title].index[0]
    
    # Get similarity scores
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sort by similarity score
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Get top_n most similar (excluding itself)
    sim_scores = sim_scores[1:top_n+1]
    
    # Return titles
    anime_indices = [i[0] for i in sim_scores]
    return df['name'].iloc[anime_indices]
```

```python
print(recommend_anime("Naruto"))
```

Q1. Difference between User-based and Item-based Collaborative Filtering

User-based: Finds users similar to the target user, then recommends items those similar users liked.

Item-based: Finds items similar to what the target user already liked, then recommends those items.

Q2. What is Collaborative Filtering and How Does it Work?

Definition: A recommendation method that makes predictions based on past behavior of many users.

How it works:

Collect user-item interactions (ratings, clicks, views).

Compute similarity either between users or items.

Recommend items that similar users liked or that are similar to what the user liked.
